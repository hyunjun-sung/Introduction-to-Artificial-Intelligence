```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix

from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.optimizers import Adam, SGD
from tensorflow.keras import backend as K
```

```python
# 데이터 불러오기
file_path = r"C:\Users\tjdgu\CodingFiles\AI_Introduction\week12\seeds_dataset.txt"
df = pd.read_csv(file_path, sep="\s+", header=None)

df
```

```python
# 컬럼명 정의
column_names = [
    "Area", "Perimeter", "Compactness", 
    "Kernel Length", "Kernel Width", 
    "Asymmetry Coefficient", "Groove Length", 
    "Class Label"
]

# 컬럼명 설정
df.columns = column_names

# 데이터 확인
df.head()
```

```python
# 결측치 확인
df.isnull().sum()
```

```python
# 'Class Label' 컬럼을 제외한 데이터를 X로 설정
X = df.drop('Class Label', axis=1).values

# 'Class Label' 컬럼만 y로 설정
y = df['Class Label'].values
```

```python
X
```

```python
y
```

```python
# 정규화
scaler = StandardScaler()
X = scaler.fit_transform(X)

# 원핫 인코딩
Y = to_categorical(y-1)
```

```python
X
```

```python
Y
```

```python
# 훈련, 테스트 데이터로 나누기 (8:2 비율)
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# 훈련, 테스트 데이터 형태 확인
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
inputs = Input(shape=(7,))  # 입력층 (7개의 feature)
x = Dense(10, activation="sigmoid")(inputs)  # 은닉층 (노드 10개, sigmoid)
prediction = Dense(3, activation="softmax")(x)  # 출력층 (클래스 3개 → softmax)

model = Model(inputs=inputs, outputs=prediction)

# 모델 컴파일
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# 모델 요약
model.summary()
```

```python
# 모델 학습
model_history = model.fit(x=X_train, y=y_train, epochs=50, batch_size=16, validation_data=(X_test, y_test))
```

```python
# 예측
y_pred = model.predict(X_test)

# 원핫 인코딩된 결과를 클래스 번호로 변환
y_test_class = np.argmax(y_test, axis=1)
y_pred_class = np.argmax(y_pred, axis=1)

# 평가 지표 출력
print(classification_report(y_test_class, y_pred_class))
print(confusion_matrix(y_test_class, y_pred_class))
```
