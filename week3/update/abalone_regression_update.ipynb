```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from scipy import stats
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
```

```python
# 데이터 불러오기

file_path = r"C:\Users\tjdgu\CodingFiles\AI_Introduction\week3\abalone.csv"
df = pd.read_csv(file_path, index_col=0)  # 첫번째 열(id)을 인덱스로 사용

df
```

```python
# 결측치 확인
df.isnull().sum()
```

```python
# 데이터 encoding
label_encoder = LabelEncoder()

# 특정 컬럼만 인코딩
df['Sex'] = label_encoder.fit_transform(df['Sex'])

df
```

```python
#상관 관계 행렬
# 0.3 이상인 경우 상관관계가 있고, 0.7이상이면 아주 높음

df_corr=df.corr()   # df의 숫자형 열들끼리 상관관계 계산

#히트맵
plt.figure(figsize=(10,10))
sns.set(font_scale=0.8)
sns.heatmap(df_corr, annot=True, cbar=False)
plt.show()

# Rings와 높은 건 Shell_weight
```

```python
# 'Rings' 컬럼을 제외한 데이터를 X로 설정
X = df.drop('Rings',axis=1)

# 'Rings' 컬럼만 y로 설정
y = df['Rings']
```

```python
X.head()
```

```python
y.head()
```

```python
# 학습 데이터(train)와 테스트 데이터(test) 8:2 비율로 분할
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.2, shuffle=True, random_state=12)

# 훈련, 테스트 데이터 형태 확인
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# 선형회기 학습, 예측
model = LinearRegression()
model.fit(X_train, y_train)
ly_preds = model.predict(X_test)
```

```python
# 선형회귀 MSE

# 함수사용
print('평균제곱근오차', mean_squared_error(ly_preds, y_test))

# MSE를 직접 구현
def mse_np(actual, predicted):
    return np.mean((np.array(actual) - np.array(predicted)) ** 2)

print('평균제곱근오차', mse_np(ly_preds, y_test))

# MSE를 for문으로 구현
def mse(actual, predicted):
    sum_square_error = sum((a - p) ** 2 for a, p in zip(actual, predicted))
    mean_square_error = sum_square_error / len(actual)
    return mean_square_error

print('평균제곱근오차', mse(ly_preds, y_test))
```

```python
# 선형회귀 예측값과 실제 값 차이를 산접도로 표시 / 노란색(예측), 파란색(정답)

plt.figure(figsize=(10,5))
plt.scatter(X_test['Shell_weight'], y_test, label='y_test') # 정답
plt.scatter(X_test['Shell_weight'], ly_preds, c='y', label='ly_preds')  # 예측
# plt.plot(X_train, lr.predict(X_train), color='blue')
plt.show()         
```

```python
# 결정트리회귀 학습, 예측
model = DecisionTreeRegressor()
model.fit(X_train, y_train)
dy_preds = model.predict(X_test)

# 결정트리회귀 MSE
print('평균제곱근오차', mean_squared_error(dy_preds, y_test))
```

```python
# 결정트리회귀 예측값과 실제 값 차이를 산접도로 표시 / 초록색(예측), 파란색(정답)
plt.figure(figsize=(10,5))
plt.scatter(X_test['Shell_weight'], y_test, label='y_test')
plt.scatter(X_test['Shell_weight'], dy_preds, c='g', label='dy_preds')

plt.show()         
```

```python
# 랜덤포레스트회귀 학습, 예측
model = RandomForestRegressor()
model.fit(X_train, y_train)
ry_preds = model.predict(X_test)

# 랜덤포레스트회귀 MSE
print('평균제곱근오차', mean_squared_error(ry_preds, y_test))
```

```python
# 랜덤포레스트회귀 예측값과 실제 값 차이를 산접도로 표시 / 주황색색(예측), 파란색(정답)
plt.figure(figsize=(10,5))
plt.scatter(X_test['Shell_weight'], y_test, label='y_test')
plt.scatter(X_test['Shell_weight'], ry_preds, c='orange', label='ry_preds')

plt.show()         
```

```python
# SVR 학습, 예측
model = SVR(kernel='linear')
model.fit(X_train, y_train)
ry_preds = model.predict(X_test)

# SVR MSE
print('평균제곱근오차', mean_squared_error(ry_preds, y_test))
```

```python
# SVR 예측값과 실제 값 차이를 산접도로 표시 / 주황색(예측), 파란색(정답)
plt.figure(figsize=(10,5))
plt.scatter(X_test['Shell_weight'], y_test, label='y_test')
plt.scatter(X_test['Shell_weight'], ry_preds, c='orange', label='ry_preds')

plt.show()    
```
